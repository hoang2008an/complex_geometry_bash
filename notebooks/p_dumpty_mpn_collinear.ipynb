# Bài toán Dumpty trong tứ giác nội tiếp

**Giả thiết**: Tứ giác `ABCD` nội tiếp, `P = AC \cap BD`.

Gọi `X, Y, Z, W` lần lượt là điểm **P-Dumpty** của các tam giác:
- `PAB` (điểm `X`)
- `PBC` (điểm `Y`)
- `PCD` (điểm `Z`)
- `PDA` (điểm `W`)

Gọi:
- `M` là trung điểm của `XZ`
- `N` là trung điểm của `YW`

Cần chứng minh: `M, P, N` thẳng hàng.

Notebook này dựng cấu hình bằng `GeometryEngine` và kiểm tra điều kiện thẳng hàng dưới dạng đa thức.

```python
from pathlib import Path
import sys
import sympy as sp

# Cho phép import khi mở notebook từ thư mục project hoặc notebooks/
root = Path.cwd()
if not (root / "src").exists() and (root.parent / "src").exists():
    root = root.parent

sys.path.insert(0, str(root / "src" / "lib"))
sys.path.insert(0, str(root / "src"))

from geometry_engine import GeometryEngine
```

```python
eng = GeometryEngine()

# Dựng tứ giác nội tiếp ABCD trên đường tròn đơn vị
for name in ["A", "B", "C", "D"]:
    eng.add_point(name)
    eng.add_unit_circle(name)

# P = AC ∩ BD
eng.intersection_of_lines("A", "C", "B", "D", "P")

# X, Y, Z, W là P-Dumpty của các tam giác tương ứng
eng.dumpty_point("P", "A", "B", "X")  # trong tam giác PAB
eng.dumpty_point("P", "B", "C", "Y")  # trong tam giác PBC
eng.dumpty_point("P", "C", "D", "Z")  # trong tam giác PCD
eng.dumpty_point("P", "D", "A", "W")  # trong tam giác PDA

# M, N là trung điểm
eng.midpoint("X", "Z", "M")
eng.midpoint("Y", "W", "N")

# In vài điểm để kiểm tra cấu hình
summary = eng.point_summary(["A", "B", "C", "D", "P", "X", "Y", "Z", "W", "M", "N"], style="text")
for k, v in summary.items():
    print(f"{k} = {v}")
```

```python
# Kiểm tra điều kiện thẳng hàng M, P, N
num, den = eng.constraint_conjugate_free("collinear", ["M", "P", "N"])[0]

num_s = sp.simplify(eng._apply_all(num))
den_s = sp.simplify(eng._apply_all(den))

print("Numerator:")
print(eng.format_expr(num_s, style="text"))
print("\nDenominator:")
print(eng.format_expr(den_s, style="text"))

if sp.simplify(num_s) == 0:
    print("\n=> Kết luận: M, P, N thẳng hàng (đa thức tử số triệt tiêu).")
else:
    print("\n=> Tử số chưa rút gọn về 0 trực tiếp. Có thể cần thêm giả thiết không suy biến hoặc đơn giản hóa sâu hơn.")
```

## Ghi chú

- Nếu kết quả `Numerator = 0`, ta đã xác nhận đẳng thức thẳng hàng trong mô hình đại số phức của engine.
- Nếu chưa về 0 ngay, có thể thử:
  - bổ sung giả thiết không suy biến (các điểm phân biệt, mẫu số khác 0),
  - hoặc chạy thêm `sp.factor(...)`, `sp.cancel(...)`, `sp.together(...)` trước khi kết luận.
- Bạn cũng có thể chuyển bài toán này thành script JSON để chạy qua `bin/geometry run ...`.

```python
from IPython.display import display, Markdown, Math

names = ["A", "B", "C", "D", "P", "X", "Y", "Z", "W", "M", "N"]
latex_summary = eng.point_summary(names, style="latex")

# Bảng Markdown với biểu thức LaTeX
rows = ["| Point | Coordinate |", "|---|---|"]
for name in names:
    rows.append(f"| ${name}$ | ${latex_summary[name]}$ |")
display(Markdown("\n".join(rows)))

# Hiển thị từng dòng công thức riêng
for name in names:
    display(Math(f"{name} = {latex_summary[name]}"))
```

```python
import sympy as sp
from IPython.display import display, Math

# Expand toạ độ M; giữ N ở dạng simplify
zM = sp.expand(sp.together(eng._apply_all(eng.z("M"))))
zbM = sp.expand(sp.together(eng._apply_all(eng.zb("M"))))
zN = sp.simplify(eng._apply_all(eng.z("N")))
zbN = sp.simplify(eng._apply_all(eng.zb("N")))

print("Coordinates (M expanded, N simplified):")
print("M =", eng.format_expr(zM, style="text"))
print("conj(M) =", eng.format_expr(zbM, style="text"))
print("N =", eng.format_expr(zN, style="text"))
print("conj(N) =", eng.format_expr(zbN, style="text"))

# In m-p/conj(m-p)
zP = sp.simplify(eng._apply_all(eng.z("P")))
zbP = sp.simplify(eng._apply_all(eng.zb("P")))

ratio_M = sp.simplify(sp.together((zM - zP) / (zbM - zbP)))
print("\n(m-p)/conj(m-p) =")
print(eng.format_expr(ratio_M, style="text"))

# Hiển thị đẹp dạng LaTeX
display(Math(r"m = " + eng.format_expr(zM, style="latex")))
display(Math(r"n = " + eng.format_expr(zN, style="latex")))
display(Math(r"\frac{m-p}{\overline{m-p}} = " + eng.format_expr(ratio_M, style="latex")))
```
