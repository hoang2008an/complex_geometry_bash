# Harmonic bundle via Viète on $(P,Q,K)$

We work with triangle $ABC$ on the unit circle, pick an arbitrary point $P$, set up the derived points
$$X = BP \cap AC, \quad Y = CP \cap AB, \quad Q = (AXY) \cap (ABC)\setminus\{A\},$$
$$K = AP \cap (ABC)\setminus\{A\}, \quad F,G = (PQK) \cap BC,$$
and verify symbolically that $(F,G;B,C) = -1$.  The notebook keeps the configuration symbolic so Viète's relations
for the two intersections with $BC$ appear directly from the quadratic obtained by eliminating the conjugate coordinate.

```python
import sys
import sympy as sp
from IPython.display import display, Math
# Make sure the engine modules are importable inside the notebook
sys.path.extend(["src", "src/lib"])
from pathlib import Path

repo_root = Path("..").resolve()         # assuming scripts/ is one level below the repo root
sys.path.extend([
    str(repo_root / "src"),
    str(repo_root / "src" / "lib"),
])


from geometry_engine import GeometryEngine

sp.init_printing()
engine = GeometryEngine()
```

```python
# Place A, B, C on the unit circle and register the main triangle scaffold.
for name in ("A", "B", "C"):
    engine.add_point(name)
    engine.add_unit_circle(name)
engine.add_point("P")
engine.point_summary(["A", "B", "C", "P"], style="text")
```

```python
# Construct X, Y, Q, K, F, G according to the problem statement.
engine.intersection_of_lines("B", "P", "A", "C", "X")
engine.intersection_of_lines("C", "P", "A", "B", "Y")

engine.circle_from_three_points("circ_AXY", "A", "X", "Y")
engine.circle_from_three_points("circ_ABC", "A", "B", "C")
engine.circle_object_intersections("circ_AXY", "circ_ABC", ["Q"], avoid=["A"])

engine.line_circle_intersection("A", "P", "O", "A", "K", avoid=["A"])
engine.circle_from_three_points("circ_PQK", "P", "Q", "K")

line_BC = engine.line_through_points("B", "C")
engine.lines["line_BC"] = line_BC
engine.line_circle_object_intersections("line_BC", "circ_PQK", ["F", "G"])

engine.point_summary(["X", "Y", "Q", "K", "F", "G"], style="text")
```

### Cross ratio check
The engine already provides a symbolic cross-ratio helper.  A quick verification shows $(F,G;B,C) = -1$.

```python
cr_FG_BC = sp.simplify(engine.cross_ratio("F", "G", "B", "C"))
cr_FG_BC, sp.simplify(cr_FG_BC )
```

```python
zF = sp.simplify(engine.z("F"))
zG = sp.simplify(engine.z("G"))
zB = sp.simplify(engine.z("B"))
zC = sp.simplify(engine.z("C"))
harmonic_nr = sp.simplify((zF - zC) * (zG - zB) + (zF - zB) * (zG - zC))
print("(F,G;B,C) + 1 ⇒", engine.format_expr(harmonic_nr, style="text"))
```

```python
engine.point_summary(["X", "Y", "Q", "K", "F", "G"], style="latex")
```

### Eliminating the conjugate: Viète's quadratic
To expose the Viète structure, eliminate $\overline{z}$ using the line equation of $BC$.
The resultant isolates a quadratic in the line parameter whose roots are $z_F$ and $z_G$.

```python
z_var = sp.Symbol("z_var")
zb_var = sp.Symbol("zb_var")

eq_line_BC = sp.expand(line_BC.alpha * z_var + line_BC.beta * zb_var + line_BC.gamma)
circle_PQK = engine.circles["circ_PQK"]
eq_circle_PQK = sp.expand(circle_PQK.evaluate(engine, z_var, zb_var))

poly_z = sp.factor(sp.resultant(eq_line_BC, eq_circle_PQK, zb_var))
poly_z
```

```python
poly_quadratic = sp.Poly(poly_z, z_var)
poly_quadratic.degree(), poly_quadratic.all_coeffs()
```

The parametrisation $\lambda = \dfrac{z - z_B}{z - z_C}$ sends $B \mapsto \infty$, $C \mapsto 0$,
so $(F,G;B,C) = -1$ is equivalent to $\lambda_F = -\lambda_G$.  Substituting $z = \dfrac{\lambda z_C - z_B}{\lambda - 1}$
into the quadratic exhibits the Viète-friendly polynomial in $\lambda$.

```python
lam = sp.Symbol("lam")
z_lam = sp.simplify((lam * engine.z("C") - engine.z("B")) / (lam - 1))
poly_lam_raw = sp.simplify(sp.together(poly_z.subs(z_var, z_lam)))
num_lam, den_lam = sp.fraction(poly_lam_raw)
poly_lam = sp.factor(num_lam)
poly_lam
```

The resulting polynomial is proportional to $\lambda^2 - 1$, so Viète tells us the two roots satisfy $\lambda_F + \lambda_G = 0$.
Consequently $\dfrac{\lambda_F}{\lambda_G} = -1$, i.e. $(F,G;B,C) = -1$, completing the harmonic bundle proof.
