```python
import sys
import sympy as sp
from IPython.display import display, Math
# Make sure the engine modules are importable inside the notebook
sys.path.extend(["src", "src/lib"])
from pathlib import Path

repo_root = Path("..").resolve()         # assuming scripts/ is one level below the repo root
sys.path.extend([
    str(repo_root / "src"),
    str(repo_root / "src" / "lib"),
])

from geometry_engine import GeometryEngine
from geometry_engine import GeometryEngine

engine = GeometryEngine()  # origin O is registered automatically

# Triangle ABC on the unit circle
for label in ("A", "B", "C"):
    engine.add_point(label)
    engine.add_unit_circle(label)

# Lemoine (symmedian) point and Fermat points
engine.lemoine_point("A", "B", "C", "L")
engine.add_fermat_points("A", "B", "C", "S", "S_alt")

# Free points and projections of P
engine.add_point("G")
engine.add_point("P")
engine.project_point_to_line("P", "B", "C", "D")
engine.project_point_to_line("P", "C", "A", "E")
engine.project_point_to_line("P", "A", "B", "F")

# Centroid of triangle DEF relabeled as G
engine.centroid("D", "E", "F", "G")




z_G = sp.simplify(engine.z("G"))
display(Math(r"z_G = " + sp.latex(z_G)))

zb_G = sp.simplify(engine.zb("G"))
display(Math(r"\bar{z}_G = " + sp.latex(zb_G)))
```
